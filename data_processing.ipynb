```python
import pandas as pd
import numpy as np
```

```python
rainfall = pd.read_excel('data/annual_rainfall.xlsx')
crop_prices = pd.read_excel('data/crop_prices.xlsx')
water_needs = pd.read_excel('data/crop_water_needs.xlsx')
crop_yields = pd.read_excel('data/crop_yields.xlsx')
```

```python
num_scen = 100
```

```python
water_needs_clean = water_needs[['Crop', 'Unnamed: 6']].copy()
water_needs_clean.columns = ['Crop', 'Water_Need_Inches']
water_needs_clean = water_needs_clean.dropna(subset=['Crop'])
water_needs_clean = water_needs_clean.dropna(subset=['Water_Need_Inches'])
water_needs_clean
```

```python
expanded_rows = []
for idx, row in water_needs_clean.iterrows():
    crops = row['Crop'].split('/')
    for c in crops:
        expanded_rows.append({'Crop': c.strip(), 'Water_Need_Inches': row['Water_Need_Inches']})
water_needs_expanded = pd.DataFrame(expanded_rows)
water_needs_expanded
```

```python
crop_prices_clean = crop_prices[['Commodity', 2025,          2024,
                2023,          2022,          2021,          2020,
                2019,          2018,          2017,          2016,
                2015,          2014,          2013,          2012,
                2011,          2010,          2009,          2008,
                2007,          2006,          2005,          2004,
                2003,          2002,          2001,          2000]].copy()
crop_prices_clean.columns = ['Crop', 2025,          2024,
                2023,          2022,          2021,          2020,
                2019,          2018,          2017,          2016,
                2015,          2014,          2013,          2012,
                2011,          2010,          2009,          2008,
                2007,          2006,          2005,          2004,
                2003,          2002,          2001,          2000]
crop_prices_clean['Crop'] = crop_prices_clean['Crop'].replace('Winter Wheat', 'Wheat')
crop_prices_clean['Crop'] = crop_prices_clean['Crop'].replace('Spring Wheat', 'Wheat')
crop_prices_clean['Crop'] = crop_prices_clean['Crop'].replace('Cotton, Upland', 'Cotton')

crop_prices_clean
```

```python
print(crop_prices_clean['Crop'].unique())
```

```python
yields_clean = crop_yields[['Commodity', 'YIELD']].copy()
yields_clean.columns = ['Crop', 'Yield']
yields_clean = yields_clean[yields_clean['Yield'].apply(lambda x: str(x).replace('.', '', 1).isdigit())]
yields_clean['Yield'] = yields_clean['Yield'].astype(float)
yields_clean
```

```python
print(yields_clean['Crop'].unique())
```

```python
def clean_crop_name(name):
    name = name.upper().strip()
    if name.endswith('S') and not name.endswith('SS') and name not in ["COTTON", "HAY"]:
        name = name[:-1]
    return name

crop_prices_clean['Crop'] = crop_prices_clean['Crop'].apply(clean_crop_name)
water_needs_expanded['Crop'] = water_needs_expanded['Crop'].apply(clean_crop_name)
yields_clean['Crop'] = yields_clean['Crop'].apply(clean_crop_name)
```

```python
print(crop_prices_clean['Crop'].unique())
print(water_needs_expanded['Crop'].unique())
print(yields_clean['Crop'].unique())
```

```python
water_needs_expanded = water_needs_expanded[water_needs_expanded['Crop'].isin(['CORN', 'SOYBEAN', 'WHEAT', 'OAT', 'BARLEY', 'PEANUT', 'HAY', 'COTTON'])]
water_needs_expanded
```

```python
rainfall_inch = rainfall['Mean Annual Rainfall Inches']
rainfall_mean = rainfall_inch.mean()
rainfall_std = rainfall_inch.std()
print("rainfall", rainfall_mean, rainfall_std)
rainfall_samples = np.random.normal(loc=rainfall_mean, scale=rainfall_std, size=num_scen)
rainfall_samples = np.clip(rainfall_samples, 0, None)
rainfall_samples
```

```python
crop_lists = ['CORN', 'SOYBEAN', 'WHEAT', 'OAT', 'BARLEY', 'PEANUT', 'HAY', 'COTTON']
crop_prices_scen = []

for crop_name in crop_lists:
    crop_prices_crop_name = crop_prices_clean[crop_prices_clean['Crop'] == crop_name]
    crop_prices_crop_name_mean = crop_prices_crop_name[[2025,          2024,
                2023,          2022,          2021,          2020,
                2019,          2018,          2017,          2016,
                2015,          2014,          2013,          2012,
                2011,          2010,          2009,          2008,
                2007,          2006,          2005,          2004,
                2003,          2002,          2001,          2000]].mean()
    crop_prices_crop_name_mean = crop_prices_crop_name_mean.mean()
    crop_prices_crop_name_std = crop_prices_crop_name[[2025,          2024,
                2023,          2022,          2021,          2020,
                2019,          2018,          2017,          2016,
                2015,          2014,          2013,          2012,
                2011,          2010,          2009,          2008,
                2007,          2006,          2005,          2004,
                2003,          2002,          2001,          2000]].std()
    crop_prices_crop_name_std = crop_prices_crop_name_std.std()
    print(crop_name, crop_prices_crop_name_mean, crop_prices_crop_name_std)

    crop_prices_crop_name_samples = np.random.normal(loc=crop_prices_crop_name_mean, scale=crop_prices_crop_name_std, size=num_scen)
    crop_prices_crop_name_samples = np.clip(crop_prices_crop_name_samples, 0, None)
    crop_prices_scen.append(crop_prices_crop_name_samples)
```

```python
crop_yields_scen = []

for crop_name in crop_lists:
    yields_crop_name = yields_clean[yields_clean['Crop'] == crop_name]
    yields_crop_name_mean = yields_crop_name['Yield'].mean()
    yields_crop_name_std = yields_crop_name['Yield'].std()
    print(crop_name, yields_crop_name_mean, yields_crop_name_std)

    yields_crop_name_samples = np.random.normal(loc=yields_crop_name_mean, scale=yields_crop_name_std, size=num_scen)
    yields_crop_name_samples = np.clip(yields_crop_name_samples, 0, None)
    crop_yields_scen.append(yields_crop_name_samples)
```

```python
final_data = []
for scen in range(1, num_scen+1):
    for index, crop_name in enumerate(crop_lists):
        final_data.append({
            'Crop': crop_name,
            'Scenario': scen,
            'Rainfall_Inches': rainfall_samples[scen-1],
            'Price ($/BU)': crop_prices_scen[index][scen-1],
            'Yield (BU/ARCE)': crop_yields_scen[index][scen-1]
        })

final_df = pd.DataFrame(final_data)
final_df.to_csv('data/final_crop_data_processed_100.csv', index=False)
```
